```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv('NYPD_Motor_Vehicle_Collisions.csv', encoding='Latin1')
df.head()
```

```python
df.columns
```

```python
working_data = df[['DATE', 'TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE', 'LONGITUDE',
       'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
       'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
       'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
       'CONTRIBUTING FACTOR VEHICLE 1', 'VEHICLE TYPE CODE 1']]
```

```python
working_data.head()
```

```python
working_data.tail()
```

```python
working_data['DATE'] = pd.to_datetime(working_data['DATE'])
```

```python
working_data['YEAR'] = working_data['DATE'].dt.year
```

```python
working_data.tail()
```

```python
df1 = working_data.groupby(working_data['YEAR']).sum().reset_index()
df1
```

```python
df1.plot(y='NUMBER OF PERSONS KILLED', x = 'YEAR', kind='barh', color = 'red', figsize = (12, 6) )

```

```python
df1.plot(y= 'NUMBER OF CYCLIST KILLED', x = 'YEAR', kind = 'line', style = '.-', color = 'pink', figsize = (12, 6))

```

```python
df1.plot(y= 'NUMBER OF PEDESTRIANS KILLED', x= 'YEAR', kind = 'area', color = 'red', style = '.', figsize= (12, 6))
```

```python
df1.plot(y ='NUMBER OF PERSONS INJURED', x = 'YEAR', kind = 'bar', color = 'red', figsize = (12, 6))
```

```python
working_data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().tail(10).plot(kind = 'barh', color = 'red', figsize= (12, 6))

plt.xlabel('10 least causes of road accidents in New York City')
```

```python
working_data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(10).plot(kind = 'barh', color= 'red', figsize= (12, 6))

plt.xlabel('10 most common causes of road accidents in New York City')
```

```python
working_data['VEHICLE TYPE CODE 1'].value_counts().head(10).plot(kind = 'barh', color= 'red', figsize = (12, 6))
```
