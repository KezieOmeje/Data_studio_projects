```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv('NYPD_Motor_Vehicle_Collisions.csv', encoding='Latin1')
df.head()
```

```python
df.columns
```

```python
working_data = df[['DATE', 'TIME', 'BOROUGH', 'ZIP CODE', 'LATITUDE', 'LONGITUDE',
       'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
       'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
       'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
       'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
       'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED',
       'CONTRIBUTING FACTOR VEHICLE 1', 'VEHICLE TYPE CODE 1']]
```

```python
working_data.head()
```

```python
working_data.tail()
```

```python
working_data['DATE'] = pd.to_datetime(working_data['DATE'])
```

```python
working_data['YEAR'] = working_data['DATE'].dt.year
```

```python
working_data.tail()
```

```python
df1 = working_data.groupby(working_data['YEAR']).sum()
df1
```

```python
df1.plot(y='NUMBER OF PERSONS KILLED', kind='barh', color = 'red')

```

```python
df1.plot(y= 'NUMBER OF CYCLIST KILLED', kind = 'bar', color = 'pink')
```

```python
df1.plot(y= 'NUMBER OF PEDESTRIANS KILLED', kind = 'barh')
```

```python
df1.plot(y ='NUMBER OF PERSONS INJURED', kind = 'bar')
```
